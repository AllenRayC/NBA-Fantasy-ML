```python
#Bring in data from internet as object then convert ot Json
schedule_response_obj=requests.get("http://data.nba.net/prod/v1/2018/schedule.json")
schedule_response=schedule_response_obj.json()
```

```python
# OPEN ITEMS!!!!!!!
# If statement to exclude if tags = "pre-season"
# tags=schedule_response_list["tags"]
#     item_dict["tags"]=item["tags"]

#Bring in elements for NBA Schedule
NBA_Schedule=[]
for item in schedule_response["league"]["standard"]:
    item_dict={}
    item_dict["gameId"]=item["gameId"]
    item_dict["startDateEastern"]=item["startDateEastern"]
    item_dict["hTeam"]=item["hTeam"]["teamId"]
    item_dict["vTeam"]=item["vTeam"]["teamId"]
    item_dict["national_broadcast"]=item["watch"]["broadcast"]["video"]["national"]["broadcasters"]
    item_dict["seasonStageId"]=item["seasonStageId"]
# Create an Object list to store NBA Game Information
    NBA_Schedule.append({"gameId":item_dict["gameId"],
                         "startDataEastern":item_dict["startDateEastern"],
                        ""}
```

```python
NBA_Schedule_DF_initial=pd.DataFrame(NBA_Schedule)
NBA_Schedule_DF_w_details=NBA_Schedule_DF_initial[["startDateEastern","gameID","vTeam","hTeam","national_broadcast","seasonStageId"]]
NBA_Schedule_DF_w_details.dtypes
```
